```python
# SCATTER PLOT IN SEABORN
```

```python
import matplotlib.pyplot as pt
import pandas as pd
import seaborn as sb

data = sb.load_dataset("tips")
print(data)
sb.scatterplot(data, x = "total_bill", y = "tip", hue = "time", palette = "dark", marker = "s") 
pt.legend(bbox_to_anchor = (0.5, 1.2))
pt.show()
```

```python
# HEATMAP IN SEABORN
```

```python
import matplotlib.pyplot as pt
import pandas as pd
import seaborn as sb

data = sb.load_dataset("tips")

gb = data.groupby("sex")["total_bill"].sum()
print(gb)
#sb.heatmap(gb)  
```

```python
import matplotlib.pyplot as pt
import pandas as pd
import seaborn as sb

data = sb.load_dataset("tips")

gb = data.groupby("sex").agg({"tip":"sum"})
#print(gb)
sb.heatmap(gb)  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6],
    'D': [6, 5, 4, 3, 2]
}
df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='hot')
plt.title("Correlation Heatmap")
plt.show()
```
